### –ó–∞–¥–∞—á–∞ 4

–ü–µ—Ä–µ–π–¥–∏—Ç–µ –Ω–∞ Google Colaboratory, —á—Ç–æ–±—ã —Ä–∞–±–æ—Ç–∞—Ç—å —Å GPU. –ó–∞–≥—Ä—É–∑–∏—Ç–µ —Ç—É–¥–∞ —ç—Ç–æ—Ç –Ω–æ—É—Ç–±—É–∫ (Upload notebook)
–í Colab –Ω–∞–∂–º–∏—Ç–µ –≤ –º–µ–Ω—é —Å–≤–µ—Ä—Ö—É: Runtime -> Change Runtime Type -> GPU, –≥–¥–µ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ —Å—Ç–æ–∏—Ç None
–û–±—ä—è–≤–∏—Ç–µ —Ç–µ–Ω–∑–æ—Ä a —Ä–∞–∑–º–µ—Ä–∞ (2, 3, 4) –∏ —Ç–µ–Ω–∑–æ—Ä b —Ä–∞–∑–º–µ—Ä–∞ (1, 8, 3) –Ω–∞ GPU, –∏–Ω–∏—Ü–∏–ª–∏–∑–∏—Ä—É–π—Ç–µ –∏—Ö —Å–ª—É—á–∞–π–Ω–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ (.uniform_())
–ó–∞—Ç–µ–º –∏–∑–º–µ–Ω–∏—Ç–µ —Ñ–æ—Ä–º—É —Ç–µ–Ω–∑–æ—Ä–∞ b, —á—Ç–æ–±—ã –æ–Ω–∞ —Å–æ–≤–ø–∞–¥–∞–ª–∞ —Å —Ñ–æ—Ä–º–æ–π —Ç–µ–Ω–∑–æ—Ä–∞ a, –ø–æ–ª—É—á–∏—Ç–µ —Ç–µ–Ω–∑–æ—Ä c
–ü–µ—Ä–µ–º–µ—Å—Ç–∏—Ç–µ c –Ω–∞ CPU, –ø–µ—Ä–µ–º–µ—Å—Ç–∏—Ç–µ a –Ω–∞ CPU
–û–±–µ—Ä–Ω–∏—Ç–µ –∏—Ö –≤ Variable()
–û–±—ä—è–≤–∏—Ç–µ —Ç–µ–Ω–∑–æ—Ä L = torch.mean((c - a) ** 2) –∏ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç L –ø–æ c ( —Ç–æ –µ—Å—Ç—å  ‚àÇùêø‚àÇùëê)

–í—ã–≤–µ–¥–∏—Ç–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç L –ø–æ c

```python
import torch

# –û–±—ä—è–≤–ª—è–µ–º —Ç–µ–Ω–∑–æ—Ä—ã a –∏ b –∏ –ø–µ—Ä–µ–º–µ—â–∞–µ–º –∏—Ö –Ω–∞ GPU
a = torch.rand(2, 3, 4, device='cuda').uniform_()
b = torch.rand(1, 8, 3, device='cuda').uniform_()

# –ú–µ–Ω—è–µ–º —Ñ–æ—Ä–º—É —Ç–µ–Ω–∑–æ—Ä–∞ b, —á—Ç–æ–±—ã –æ–Ω–∞ —Å–æ–≤–ø–∞–¥–∞–ª–∞ —Å —Ñ–æ—Ä–º–æ–π —Ç–µ–Ω–∑–æ—Ä–∞ a
b = b.view(2, 3, 4)

# –ü–µ—Ä–µ–º–µ—â–∞–µ–º —Ç–µ–Ω–∑–æ—Ä—ã a –∏ b –Ω–∞ CPU –∏ –æ–±–æ—Ä–∞—á–∏–≤–∞–µ–º –∏—Ö –≤ Variable
a = a.cpu()
b = b.cpu()
a.requires_grad = True
b.requires_grad = True

# –û–±—ä—è–≤–ª—è–µ–º —Ç–µ–Ω–∑–æ—Ä c
c = b

# –û–±—ä—è–≤–ª—è–µ–º —Ç–µ–Ω–∑–æ—Ä L
L = torch.mean((c - a) ** 2)

# –í—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç L –ø–æ c
L.backward()

# –í—ã–≤–æ–¥–∏–º –≥—Ä–∞–¥–∏–µ–Ω—Ç L –ø–æ c
print("–ì—Ä–∞–¥–∏–µ–Ω—Ç L –ø–æ c:")
print(c.grad)
```

### –ó–∞–¥–∞—á–∞ 5

–î–æ—Ä–∞–±–æ—Ç–∞–π—Ç–µ –º–Ω–æ–≥–æ—Å–ª–æ–π–Ω—É—é –Ω–µ–π—Ä–æ–Ω–Ω—É—é —Å–µ—Ç—å –∏–∑ —Ñ–∞–π–ª–∞ "6_3 mlp pytorch tasks".

```python
import matplotlib.pyplot as plt
import numpy as np

import torch

N = 100
D = 2
K = 3
X = np.zeros((N * K, D))
y = np.zeros(N * K, dtype='uint8')

for j in range(K):
    ix = range(N * j,N * (j + 1))
    r = np.linspace(0.0, 1, N)
    t = np.linspace(j * 4, (j + 1) * 4,N) + np.random.randn(N) * 0.2 # theta
    X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
    y[ix] = j

X = torch.autograd.Variable(torch.FloatTensor(X))
y = torch.autograd.Variable(torch.LongTensor(y.astype(np.int64)))

print(X.data.shape, y.data.shape)

# N - —Ä–∞–∑–º–µ—Ä –±–∞—Ç—á–∞ (batch_size, –Ω—É–∂–Ω–æ –¥–ª—è –º–µ—Ç–æ–¥–∞ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏);
# D_in - —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –≤—Ö–æ–¥–∞ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —É –æ–±—ä–µ–∫—Ç–∞);
# H - —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å —Å–∫—Ä—ã—Ç—ã—Ö —Å–ª–æ—ë–≤;
# D_out - —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–Ω–æ–≥–æ —Å–ª–æ—è (—Å—É—Ç—å - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–æ–≤)
N, D_in, H, D_out = 64, 2, 100, 3

# Use the nn package to define our model and loss function.
two_layer_net = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out),
)

loss_fn = torch.nn.CrossEntropyLoss(size_average=False)

learning_rate = 1e-4
optimizer = torch.optim.SGD(two_layer_net.parameters(), lr=learning_rate)
for t in range(500):
    # forward
    y_pred = two_layer_net(X)

    # loss
    loss = loss_fn(y_pred, y)
    print('{} {}'.format(t, loss.data))

    # –∑–∞–Ω—É–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç—ã (—á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –æ—Å—Ç–∞—Ç–∫–∞ —Å —Ä–µ–¥—ã–¥—É—â–µ–≥–æ —à–∞–≥–∞)
    optimizer.zero_grad()

    # backward
    loss.backward()

    # –æ–±–Ω–æ–≤–ª—è–µ–º
    optimizer.step()


X = X.data.numpy()
y = y.data.numpy()



h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
grid_tensor = torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()])

Z = two_layer_net(torch.autograd.Variable(grid_tensor))
Z = Z.data.numpy()
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))

plt.contourf(xx, yy, Z, cmap=plt.cm.rainbow, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.rainbow)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.title('–°–ø–∏—Ä–∞–ª—å', fontsize=15)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();

# –æ–±–æ—Ä–∞—á–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ
X = torch.FloatTensor(X)
y = torch.LongTensor(y)

N, D_in, H, D_out = 64, 2, 100, 3

#—Å–æ–∑–¥–∞–µ–º —Å–µ—Ç—å better_net —Å 4-–º—è —Å–ª–æ—è–º–∏
better_net = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out),
)

# —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å –∫–∞–∫ –∏ —Ä–∞–Ω—å—à–µ –ö—Ä–æ—Å—Å—ç–Ω—Ç—Ä–æ–ø–∏—è
loss_fn = torch.nn.CrossEntropyLoss(size_average=False)


# learning rate –∑–∞–¥–∞—ë–º 1e-3
learning_rate = 1e-3

#–æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä –Ω–µ –º–µ–Ω—è–µ–º
optimizer = torch.optim.SGD(better_net.parameters(), lr=learning_rate)

# –æ–±—É—á–∞–µ–º –Ω–∞ 1000 —ç–ø–æ—Ö–∞—Ö
for t in range(1000):
    # forward
    y_pred = better_net(X)

    # loss
    loss = loss_fn(y_pred, y)
    print('{} {}'.format(t, loss.data))

    # –∑–∞–Ω—É–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç—ã
    optimizer.zero_grad()

    # backward
    loss.backward()

    # –æ–±–Ω–æ–≤–ª—è–µ–º
    optimizer.step()

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
grid_tensor = torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()])

Z = better_net(torch.autograd.Variable(grid_tensor))
Z = Z.data.numpy()
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))

plt.contourf(xx, yy, Z, cmap=plt.cm.rainbow, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.rainbow)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.title('–°–ø–∏—Ä–∞–ª—å', fontsize=15)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();
```
